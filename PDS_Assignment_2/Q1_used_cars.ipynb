```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv")   # if that is the correct file
df.head()
```

```python
# Convert numeric columns containing units into pure numbers

# Mileage: "18.5 kmpl" or "13 km/kg" -> 18.5, 13
if 'Mileage' in df.columns:
    df['Mileage'] = (
        df['Mileage']
        .astype(str)
        .str.extract(r'([\d\.]+)')
        .astype(float)
    )

# Engine: "1197 CC" -> 1197
if 'Engine' in df.columns:
    df['Engine'] = (
        df['Engine']
        .astype(str)
        .str.extract(r'([\d\.]+)')
        .astype(float)
    )

# Power: "74 bhp" or "null bhp" -> 74
if 'Power' in df.columns:
    df['Power'] = (
        df['Power']
        .astype(str)
        .str.extract(r'([\d\.]+)')
        .astype(float)
    )

# New Price: "5.34 Lakh" -> 5.34
if 'New_Price' in df.columns:
    df['New_Price'] = (
        df['New_Price']
        .astype(str)
        .str.extract(r'([\d\.]+)')
        .astype(float)
    )
```

```python
num_cols = ['Mileage', 'Engine', 'Power', 'New_Price', 'Kilometers_Driven']

for col in num_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())
```

```python
cat_cols = ['Fuel_Type', 'Transmission']

for col in cat_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mode()[0])
```

```python
# ----- ONE HOT ENCODING -----

cols_to_encode = [col for col in ['Fuel_Type', 'Transmission'] if col in df.columns]

df_encoded = pd.get_dummies(df, columns=cols_to_encode, drop_first=True)

df_encoded.head()
```

```python
# ----- CREATE ONE NEW FEATURE -----

current_year = 2024

if 'Year' in df_encoded.columns:
    df_encoded['Car_Age'] = current_year - df_encoded['Year']

df_encoded[['Year', 'Car_Age']].head()
```

```python
# ----- SELECT -----
df_select = df_encoded[['Price', 'Mileage', 'Engine', 'Car_Age']]
df_select.head()
```

```python
# ----- FILTER -----
df_filter = df_encoded[df_encoded['Car_Age'] < 10]
df_filter.head()
```

```python
# ----- RENAME -----
df_rename = df_encoded.rename(columns={'Price': 'Used_Car_Price'})
df_rename[['Used_Car_Price']].head()
```

```python
# ----- MUTATE (create new column) -----
df_encoded['Price_per_km'] = df_encoded['Price'] / df_encoded['Kilometers_Driven']
df_encoded[['Price', 'Kilometers_Driven', 'Price_per_km']].head()
```

```python
# ----- ARRANGE (sort) -----
df_sorted = df_encoded.sort_values(by='Price', ascending=False)
df_sorted[['Name', 'Price']].head()   # if Name column exists
```

```python
# ----- SUMMARISE + GROUPBY -----
if 'Fuel_Type_Petrol' in df_encoded.columns:
    print(df_encoded.groupby('Fuel_Type_Petrol')['Price'].mean())
```

```python
plt.hist(df_encoded['Price'], bins=30)
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Distribution of Used Car Prices')
plt.show()
```
